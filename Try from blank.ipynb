```python
#Imports stuff
import matplotlib.pyplot as plt
import numpy as np
from astropy.visualization import astropy_mpl_style
plt.style.use(astropy_mpl_style)
import glob
from astropy.utils.data import get_pkg_data_filename
from astropy.io import fits

def get_data(n):
#sets up directory and makes a couple variables
    fpath = '/Users/fixit/Desktop/Python/Research/LTT1445_fits_files/'
    list = glob.glob(fpath+'*.fits')
    list = sorted(list)
#displays image info
    image_file = get_pkg_data_filename(list[n])
    fits.info(image_file)
#shows image
    image_data = fits.getdata(image_file, ext=0)
    print(image_data.shape)
    plt.figure()
    plt.imshow(image_data, cmap='gray')
    plt.colorbar()

# gets timestamps and prints
    hdul = fits.open(image_file)
    expstart = hdul[0].header['EXPSTART']
    expend = hdul[0].header['EXPEND']
    exptime = hdul[0].header['EXPTIME']
    timestamp = np.median([expstart, expend])
    print("___________________________________________")        
    print("Timestamp:", timestamp)
#Finds background value and prints
    x = np.zeros(len(image_data))
    for n in range(len(image_data)):
        x[n] = np.sum(image_data[n][:])
    background = np.median(x)
    print("___________________________________________")        
    print("Background:", background)
```
